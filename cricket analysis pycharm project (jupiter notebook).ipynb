```python
def selectoptiondf(df,name,year_of_match,month_of_match,matchtype,run_rate):
    
    
    if name == 'Overall' and year_of_match == 'Overall' and month_of_match == 'Overall' and matchtype == 'Overall' and run_rate == 'Overall':
        tem = df
        
    if name != 'Overall' and year_of_match == 'Overall' and month_of_match == 'Overall' and matchtype == 'Overall' and run_rate == 'Overall':
        tem = df[(df['name'] == name)]
        
    if name == 'Overall' and year_of_match != 'Overall' and month_of_match == 'Overall' and matchtype == 'Overall' and run_rate == 'Overall':
        tem = df[(df['year_of_match'] == year_of_match)]
        
    if name == 'Overall' and year_of_match == 'Overall' and month_of_match != 'Overall' and matchtype == 'Overall' and run_rate == 'Overall':
        tem = df[(df['month_of_match'] == month_of_match)]
        
    if name == 'Overall' and year_of_match == 'Overall' and month_of_match == 'Overall' and matchtype != 'Overall' and run_rate == 'Overall':
        tem = df[(df['matchtype'] == matchtype)]
     
    if name == 'Overall' and year_of_match == 'Overall' and month_of_match == 'Overall' and matchtype == 'Overall' and run_rate != 'Overall':
        tem = df[(df['run_rate'] == run_rate)]
        
    #1    
    if name != 'Overall' and year_of_match != 'Overall' and month_of_match == 'Overall' and matchtype == 'Overall' and run_rate == 'Overall':
        tem = df[(df['name'] == name) & (df['year_of_match'] == year_of_match)]
        
        
    if name != 'Overall' and year_of_match == 'Overall' and month_of_match != 'Overall' and matchtype == 'Overall' and run_rate == 'Overall':
        tem = df[(df['name'] == name) & (df['month_of_match'] == month_of_match)]
     
       
    if name != 'Overall' and year_of_match == 'Overall' and month_of_match == 'Overall' and matchtype != 'Overall' and run_rate == 'Overall':
        tem = df[(df['name'] == name) & (df['matchtype'] == matchtype)] 
     
    if name != 'Overall' and year_of_match == 'Overall' and month_of_match == 'Overall' and matchtype == 'Overall' and run_rate != 'Overall':
        tem = df[(df['name'] == name) & (df['run_rate'] == run_rate)]
        
      #conect 3 
    
    if name != 'Overall' and year_of_match != 'Overall' and month_of_match != 'Overall' and matchtype == 'Overall' and run_rate == 'Overall':
        tem = df[(df['name'] == name) & (df['year_of_match'] == year_of_match) & (df['month_of_match'] == month_of_match)]   
        
        
    if name != 'Overall' and year_of_match != 'Overall' and month_of_match == 'Overall' and matchtype != 'Overall' and run_rate == 'Overall':
         tem = df[(df['name'] == name) & (df['year_of_match'] == year_of_match) & (df['matchtype'] == matchtype)]
        
    if name != 'Overall' and year_of_match != 'Overall' and month_of_match == 'Overall' and matchtype == 'Overall' and run_rate != 'Overall':
         tem = df[(df['name'] == name) & (df['year_of_match'] == year_of_match) & (df['run_rate'] == run_rate)]
    
    #connect 4
    
    if name != 'Overall' and year_of_match != 'Overall' and month_of_match != 'Overall' and matchtype != 'Overall' and run_rate == 'Overall':
        tem = df[(df['name'] == name) & (df['year_of_match'] == year_of_match) & (df['month_of_match'] == month_of_match) & (df['matchtype'] == matchtype)]  
        
        
    if name == 'Overall' and year_of_match != 'Overall' and month_of_match != 'Overall' and matchtype == 'Overall' and run_rate != 'Overall':
        tem = df[(df['name'] == name) & (df['year_of_match'] == year_of_match) & (df['month_of_match'] == month_of_match) & (df['run_rate'] == run_rate)]  
      
    #5connect
    if name != 'Overall' and year_of_match != 'Overall' and month_of_match != 'Overall' and matchtype != 'Overall' and run_rate != 'Overall':
        tem = df[(df['name'] == name) & (df['year_of_match'] == year_of_match) & (df['month_of_match'] == month_of_match) &(df['matchtype'] == matchtype) & (df['run_rate'] == run_rate)]  
        
        
    
 
    x=tem['name'].value_counts().reset_index()
    x.columns=['country','paticipate matchas']
   
    x1=tem['year_of_match'].value_counts().reset_index()
    x1.columns=['year','paticipate matchas']
    
    x2=tem['month_of_match'].value_counts().reset_index()
    x2.columns=['Month','paticipate matchas']
    
    x3=tem['matchtype'].value_counts().reset_index()
    x3.columns=['matchtype','paticipate matchas']
 
    x4=tem['run_rate'].value_counts().reset_index()
    x4.columns=['run rate','counts of runrate']
    
  
    
    
    print(x)  
    print(x1)
    print(x2)
    print(x3)
    print(x4)
```
